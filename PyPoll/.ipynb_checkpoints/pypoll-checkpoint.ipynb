```python
# Dependencies
import csv
import os

# Files to load
election_data_csv = os.path.join("Resources", "election_data.csv")
# Specify the file to write to
analysis_printed = os.path.join("analysis", "Analysis_PyPoll.txt")
```

```python
# create lists to append to and counters

#counter to add the total number of votes cast
votes = 0

#list of candidates and dictionary of how many votes each received
candidates = []
candidate_votes = {}

# Winning Candidate tracker and Winning vote counter
winning_candidate = ""
winning_votes = 0

# Read the csv and convert it into a list of dictionaries
with open(election_data_csv) as election_data:
    reader = csv.reader(election_data)

    # Skip the header
    header = next(reader)

    # For loop through reader to gather data for analysis
    for row in reader:

        # Add 1 to vote count
        votes = votes + 1

        # Extract the candidate name from each row
        candidate = row[2]

        # If the candidate does is not yet in candiates list append their name to this list
        if candidate not in candidates:
            candidates.append(candidate)

            # start tracking votes for the added candidate and add a vote to their vote count
            candidate_votes[candidate] = 0
        candidate_votes[candidate] = candidate_votes[candidate] + 1

# Print the results and export the data as txt file
with open(analysis_printed, "w") as txt_file:

    # Set style for printing the total votes (this is for both showing in notebook and printing to txt)
    election_results = (
        f"\n\nElection Results\n"
        f"-------------------------\n"
        f"Total Votes: {votes}\n"
        f"-------------------------\n")
    print(election_results, end="")

    # Save the styled election results (above) text file
    txt_file.write(election_results)

    # Next determine the winner by looping through the candiate votes dictionary (created above)
    for candidate_name in candidate_votes:

        # Retrieve candiate voted for as number of votes and calculate percentage of votes
        cand_voted = candidate_votes.get(candidate_name)
        vote_percentage = float(cand_voted) / float(votes) * 100

        # Determine winning vote count and candidate
        if (cand_voted > winning_votes):
            winning_votes = cand_voted
            winning_candidate = candidate_name

        # set each candidate's votes and percentage for printing
        votes_output = f"{candidate_name}: {vote_percentage:.3f}% ({cand_voted})\n"
        print(votes_output, end="")

        # Save to text file the votes output (set above)
        txt_file.write(votes_output)

    # Set the style of printing for winning candidate (for notebook and to save to text file)
    winning_candidate_summary = (
        f"-------------------------\n"
        f"Winner: {winning_candidate}\n"
        f"-------------------------\n")
    print(winning_candidate_summary)

    # Save the winning candidate's to the text file
    txt_file.write(winning_candidate_summary)

```
